## Karatsuba for Multiplication using dynamic programming

```python
import numpy as np
```

```python
# function to split the number into 2 halves
def split(number, m2):
    num_text = str(number)
    high = int(num_text[:-m2])
    low = int(num_text[-m2:])
    return low, high

# main function
def Karatsuba(num1, num2):
    # base case
    if num1 < 10 or num2 < 10:
        return num1 * num2
    else:
        m = max(len(str(num1)), len(str(num2)))
        m2 = m // 2
        low1, high1 = split(num1, m2)
        low2, high2 = split(num2, m2)
        c0 = Karatsuba(high1, high2)
        c1 = Karatsuba((low1 + high1), (low2 + high2))
        c2 = Karatsuba(low1, low2)
        return ((c0 * (10**(m2 * 2))) + (c1 - c0 - c2) * (10**m2) + c2)
```

## Testing the code

```python
print(np.allclose(Karatsuba(11,22), 11 * 22))
```

```python
print(np.allclose(Karatsuba(121,222), 121 * 222))
```

```python
print(np.allclose(Karatsuba(1231,2223), 1231 * 2223))
```
